```python
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2, ds_tfms=get_transforms(), size=128, num_workers=4).normalize(imagenet_stats)
learn = cnn_learner(data, models.resnet18, metrics = error_rate)
```

```python
learn.lr_find()
learn.recorder.plot(return_fig=True).savefig("lrfind.jpg")
```

```python
lr  = 0.01
learn.fit_one_cycle(1, slice(lr))
```

Save the weights

```python
dest = "Resnet18-128-stage-1"
learn.save(dest)
```

Unfreeze the outer section of the network

```python
learn.unfreeze()
```

Find an appropriate learning rate

```python
learn.lr_find()
learn.recorder.plot(return_fig=True).savefig("lrfind.jpg")
```

```python
learn.fit_one_cycle(5, slice(1e-4,lr/5))
```

Save the weights

```python
dest = "Resnet18-128-stage-2"
learn.save(dest)
```
